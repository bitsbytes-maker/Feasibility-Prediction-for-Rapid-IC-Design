```
#Six curves for five variables matrix (10x5).... Max vs Min
from numpy import random
import matplotlib.pyplot as plt 
import numpy as np


def getMaxIndex(x):
  indexFirst = 0 
  indexSecond = 0
  max = x[0]
  for i in range(len(x)):
    if max <= x[i]:
      max = x[i]
      indexFirst = i
  
  x[indexFirst] = -max*max
  max = x[0]

  for i in range(len(x)):
    if max <= x[i]:
      max = x[i]
      indexSecond = i



  return indexFirst,indexSecond 

def squr(x,y):
  return x*x - y*y

def drawCurve(x,y):
  plt.figure(figsize=(10,10))
  #plt.scatter(x,y, s= 20) # for changing it to scatter plt.scatter() and keep the s which is point size
  plt.plot(x,y)  # for changing it to striaght line plt.plot() and omit the s which is point size
  for i in range(len(x)):
    plt.annotate(str(i), (x[i],y[i]), size = 20)
  plt.xlabel(str(x))
  plt.ylabel(str(y))

x = random.randint(100, size=(10, 5))
print(x)

# x for maximizing and  y for minimizing
x0,x1,x2,y3,y4 = [],[],[],[],[]

#0,1..4 that's are coulumn number
o1,o2,o3,o4,o5,o6 = [],[],[],[],[],[]

#It is the priority
a,b,c,d,e = 1,1,1,1,1

for i in range(len(x)):
  x0.append(a*x[i][0])
  x1.append(b*x[i][1])
  x2.append(c*x[i][2])
  y3.append(d*x[i][3])
  y4.append(e*x[i][4])


#Just keeping the value in array for furture use when data size is big or graph number will increase
array = np.arange(60).reshape(6,10)


#array and o1,o2,o3,o4,o5,o6 are same task doing
for i in range(len(x)):
  o1.append(squr(x0[i],y3[i]))
  array[0][i] = squr(x0[i],y3[i])
  o2.append(squr(x1[i],y3[i]))
  array[1][i] = squr(x1[i],y3[i])
  o3.append(squr(x2[i],y3[i]))
  array[2][i] = squr(x2[i],y3[i])
  o4.append(squr(x0[i],y4[i])) 
  array[3][i] = squr(x0[i],y4[i])
  o5.append(squr(x1[i],y4[i])) 
  array[4][i] = squr(x1[i],y4[i])
  o6.append(squr(x2[i],y4[i]))
  array[5][i] = squr(x2[i],y4[i])


# sList stores the solution list for counting the maximum occurence 79 to 93
sList = []
graphwithPriorityDictionary = np.arange(12).reshape(6,2)
for i in range(6):
  one,two = getMaxIndex(array[i])
  graphwithPriorityDictionary[i][0] = one
  graphwithPriorityDictionary[i][1] = two
  sList.append(one)
  


print(array)


print('Graph List with Priority')
print(graphwithPriorityDictionary)


maxCount = []
for i in range(len(sList)):
  maxCount.append(sList.count(sList[i]))

maxOcurring = np.max(maxCount)
print("Our best solution may be")
print(sList[maxCount.index(maxOcurring)])
  


print('From the list')
print(sList)


print(o1)
print(o2)
print(o3)
print(o4)
print(o5)
print(o6)

print(str(getMaxIndex(o1))+"  "+str(getMaxIndex(o2))+" "+str(getMaxIndex(o3))+" "+
      str(getMaxIndex(o4))+" "+str(getMaxIndex(o5))+" "+str(getMaxIndex(o6)))


smplNo = [i for i in range(len(x))]

# drawCurve(smplNo,o1)
# drawCurve(smplNo,o2)
# drawCurve(smplNo,o3)
# drawCurve(smplNo,o4)
# drawCurve(smplNo,o5)
# drawCurve(smplNo,o6)

```
