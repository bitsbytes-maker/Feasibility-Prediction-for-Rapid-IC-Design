```
#MAtrix intialization
x = random.randint(100, size=(10, 10))
print(x)
```

```
#Objective function = Rmax - Rmin for 5 variables

import matplotlib.pyplot as plt 
from numpy import random

def squrroot(x,y,z=0):
  return (x**2+y**2+z**2)**0.5


print(x)


front = x[0]
print("Square root")

maxList = []
minList = []

# increasing the value of coeffiecen to maximize and decreasing to minimize

for i in range(len(x)):
  squr_root_max = squrroot(x[i][0],x[i][1],x[i][2])

  squr_root_min = squrroot(x[i][3],x[i][4])

  maxList.append(squr_root_max)
  minList.append(squr_root_min)

  print(str(i)+"  no row----->"+str(squrroot(x[i][0],x[i][1],x[i][2]))+"---->"+str(squrroot(x[i][3],x[i][4])))

index = 0
frontMax = maxList[0]
frontMin = minList[0]
for i in range(len(x)):
  
  if (maxList[i]- minList[i]) >= (frontMax - frontMin):
    frontMax = maxList[i]
    frontMin = minList[i]
    index = i

print('Approximate solution')
print(str(index)+" ..........is the soultion number")

plt.figure(figsize=(10,10))
plt.scatter(maxList, minList, s= 100)
for i in range(len(x)):
  plt.annotate(str(i), (maxList[i],minList[i]), size = 20)

plt.xlabel('MaxList')
plt.ylabel('MinList')


diffList = []
sampleNoList = [i for i in range(len(x))]
for i in range(len(x)):
  diffList.append(maxList[i] - minList[i])


plt.figure(figsize=(10,10))
plt.plot(sampleNoList, diffList)
for i in range(len(x)):
  plt.annotate(str(i), (sampleNoList[i],diffList[i]), size = 20)

plt.xlabel('Sample no')
plt.ylabel('Difference')
```

```
#Objective function = Rmax - Rmin having 10 variables with coefficient
import matplotlib.pyplot as plt 
from numpy import random

def squrRoot(x,y,z=0):
  return (x**2+y**2+z**2)**0.5


def optimalSolution(maxList,minList):
  index = 0
  frontMax = maxList[0]
  frontMin = minList[0]
  for i in range(len(x)):
  
    if (maxList[i]- minList[i]) >= (frontMax - frontMin):
      frontMax = maxList[i]
      frontMin = minList[i]
      index = i
  
  return index


print(x)


print("Square root with factoring")

maxList = []
minList = []

a,b,c,d = 2,3,0.9,0.9
# increasing the value of coeffiecent to maximize function and decreasing to minimize function

for i in range(len(x)):
  _max = a*squrRoot(x[i][0],x[i][1],x[i][2])+b*squrRoot(x[i][3],x[i][4],x[i][5])

  _min = c*squrRoot(x[i][6],x[i][7])+ d*squrRoot(x[i][8],x[i][9])

  maxList.append(_max)
  minList.append(_min)

  print(str(i)+"  no row----->"+str(_max)+ "---->" +str(_min))

print('Solution with factoring')
index = optimalSolution(maxList,minList)
print(str(index)+" ..........is the soultion number")

print('Without factoring square root')

maxList1 = []
minList1 = []
a,b,c,d =1,1,1,1
for i in range(len(x)):
  _max = a*squrRoot(x[i][0],x[i][1],x[i][2])+b*squrRoot(x[i][3],x[i][4],x[i][5])

  _min =c*squrRoot(x[i][6],x[i][7])+ d*squrRoot(x[i][8],x[i][9])

  maxList1.append(_max)
  minList1.append(_min)


  print(str(i)+"  no row----->"+str(_max)+"---->"+str(_min))

print('Solution without factoring')
index = optimalSolution(maxList1,minList1)
print(str(index)+" ..........is the soultion number")



plt.figure(figsize=(10,10))
plt.scatter(maxList, minList, s= 100)
for i in range(len(x)):
  plt.annotate(str(i), (maxList[i],minList[i]), size = 20)

plt.xlabel('MaxList')
plt.ylabel('MinList')


diffList = []
sampleNoList = [i for i in range(len(x))]
for i in range(len(x)):
  diffList.append(maxList[i] - minList[i])


plt.figure(figsize=(10,10))
plt.plot(sampleNoList, diffList)
for i in range(len(x)):
  plt.annotate(str(i), (sampleNoList[i],diffList[i]), size = 20)

plt.xlabel('Sample no')
plt.ylabel('Difference')
```

```
#Six curves for five variables matrix (10x5).... Max vs Min

import matplotlib.pyplot as plt 

def drawCurve(x,y):
  plt.figure(figsize=(10,10))
  plt.scatter(x,y, s= 20) # for changing it to scatter plt.scatter() and keep the s which is point size
  #plt.plot(x,y)  # for changing it to striaght line plt.plot() and omit the s which is point size
  for i in range(len(x)):
    plt.annotate(str(i), (x[i],y[i]), size = 20)
  plt.xlabel(str(x))
  plt.ylabel(str(y))



x0,x1,x2,y3,y4 = [],[],[],[],[]

#0,1..4 that's are coulumn number

for i in range(len(x)):
  x0.append(x[i][0])
  x1.append(x[i][1])
  x2.append(x[i][2])
  y3.append(x[i][3])
  y4.append(x[i][4])

 

drawCurve(x0,y3)
drawCurve(x1,y3)
drawCurve(x2,y3)
drawCurve(x0,y4)
drawCurve(x1,y4)
drawCurve(x2,y4)
```
